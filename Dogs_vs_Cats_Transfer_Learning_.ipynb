```python
import os
from PIL import Image

original_folder = '/content/train_data/train'
resized_folder = '/content/image_resized'
os.makedirs(resized_folder, exist_ok=True)

for i, filename in enumerate(os.listdir(original_folder)):
    if i == 2500:  # Stop after 2500 images
        break

    try:
        img = Image.open(os.path.join(original_folder, filename)).resize((224, 224)).convert('RGB')
        img.save(os.path.join(resized_folder, filename))
        print(f"Processed: {filename}")
    except Exception as e:
        print(f"Error processing {filename}: {e}")
```

```python
import glob
import numpy as np
import cv2

image_directory = "/content/image_resized"
image_extensions = ["png", "jpg"]

# Collect all image file paths (limit to 2000)
files = [file for ext in image_extensions for file in glob.glob(f"{image_directory}/*.{ext}")]

# Load images as a NumPy array
dog_cat_images = np.array([cv2.imread(file) for file in files])

print(f"Loaded {len(dog_cat_images)} images.")
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

#Scaling

```python
# scaling the data
X_train_scaled = X_train/255

X_test_scaled = X_test/255
```

**Building the Neural Network**

```python
import tensorflow as tf
from tensorflow import keras

```

```python
from tensorflow.keras import Sequential, models, layers
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import BatchNormalization #scaling
from tensorflow.keras.models import load_model
from tensorflow.keras.models import Model
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras import optimizers
```

```python
convolutional_base = ResNet50(weights='imagenet', include_top=False, input_shape=(224,224,3))
convolutional_base.summary()
```

```python
num_of_classes = 2

model = models.Sequential()
model.add(convolutional_base)   # Add ResNet50
model.add(layers.Flatten())
model.add(layers.BatchNormalization())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.BatchNormalization())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.BatchNormalization())
model.add(layers.Dense(num_of_classes, activation='softmax'))
```

```python
model.compile(optimizer=optimizers.RMSprop(), loss='sparse_categorical_crossentropy', metrics=['acc'])
```

```python
history = model.fit(X_train_scaled, Y_train, validation_split=0.1, epochs=10)
```

```python
loss, accuracy = model.evaluate(X_test_scaled, Y_test)
print('Test Accuracy =', accuracy)
```

```python
h= history

# plot the loss value
plt.plot(h.history['loss'], label='train loss')
plt.plot(h.history['val_loss'], label='validation loss')
plt.legend()
plt.show()

# plot the accuracy value
plt.plot(h.history['acc'], label='train accuracy')
plt.plot(h.history['val_acc'], label='validation accuracy')
plt.legend()
plt.show()
```

Predictive System

```python
input_image_path = input('Path of the image to be predicted: ')

input_image = cv2.imread(input_image_path)

cv2_imshow(input_image)

input_image_resize = cv2.resize(input_image, (224,224))

input_image_scaled = input_image_resize/255

image_reshaped = np.reshape(input_image_scaled, [1,224,224,3])

input_prediction = model.predict(image_reshaped)

print(input_prediction)

input_pred_label = np.argmax(input_prediction)

print(input_pred_label)

if input_pred_label == 0:
  print('The image represents a Cat')

else:
  print('The image represents a Dog')
```
